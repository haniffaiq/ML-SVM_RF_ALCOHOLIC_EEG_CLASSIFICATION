```python
import cv2
import matplotlib.pyplot as plt
import os


# Daftar path ke 4 gambar
image_paths = ['./Data_Enhance/Alcoholic_7.png', './Data_Enhance/Alcoholic_33.png', './Data_Enhance/Non_Alcoholic_727.png', './Data_Enhance/Non_Alcoholic_777.png']

# Loop melalui setiap path gambar
for path in image_paths:
    # Ambil nama file dari path
    file_name = os.path.basename(path)

    # Baca gambar
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # Tampilkan gambar
    plt.imshow(image, cmap='gray')
    plt.title('Gambar Skala Abu-abu - ' + file_name)
    plt.axis('off')
    plt.show()


# Loop melalui setiap path gambar
for path in image_paths:
    # Ambil nama file dari path
    file_name = os.path.basename(path)

    # Baca gambar
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)


    # Tampilkan histogram
    plt.hist(image.ravel(), bins=256, color='blue', alpha=0.7)
    plt.title('Histogram - ' + file_name)
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Jumlah Piksel')
    plt.show()
```
