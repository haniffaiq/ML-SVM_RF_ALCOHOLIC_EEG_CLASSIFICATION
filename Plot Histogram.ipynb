```python
import cv2
import matplotlib.pyplot as plt
import os


# Daftar path ke 4 gambar
image_paths = ['./Data_Enhance/Alcoholic_7.png', './Data_Enhance/Alcoholic_33.png', './Data_Enhance/Non_Alcoholic_727.png', './Data_Enhance/Non_Alcoholic_777.png']

# Tentukan ukuran subplot
num_rows = 2
num_cols = 2

# Buat subplot
fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 10))

# Loop melalui setiap path gambar
for i, path in enumerate(image_paths):
    # Ambil nama file dari path
    file_name = os.path.basename(path)

    # Baca gambar
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # Hitung indeks subplot
    row = i // num_cols
    col = i % num_cols

    # Tampilkan gambar dalam subplot yang sesuai
    axs[row, col].imshow(image, cmap='gray')
    axs[row, col].set_title('Gambar Skala Abu-abu - ' + file_name)
    axs[row, col].axis('off')

# Tampilkan gambar
plt.tight_layout()
plt.show()

# Buat subplot histogram
fig, axs = plt.subplots(num_rows, num_cols, figsize=(10, 10))

# Loop melalui setiap path gambar
for i, path in enumerate(image_paths):
    # Ambil nama file dari path
    file_name = os.path.basename(path)

    # Baca gambar
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # Hitung indeks subplot
    row = i // num_cols
    col = i % num_cols

    # Tampilkan histogram dalam subplot yang sesuai
    axs[row, col].hist(image.ravel(), bins=256, color='blue', alpha=0.7)
    axs[row, col].set_title('Histogram - ' + file_name)
    axs[row, col].set_xlabel('Intensitas Piksel')
    axs[row, col].set_ylabel('Jumlah Piksel')

# Tampilkan histogram
plt.tight_layout()
plt.show()
```
